```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

# 加载 CSV 文件
df = pd.read_csv("./data/FM/fashion-mnist_test.csv")

# 选一个样本 index（可以改）
idx = 5

# 取标签 + 像素
row = df.iloc[idx].values
label = row[0]
pixels = row[1:].reshape(28, 28)

# Fashion-MNIST 类别名称
label_names = [
    "T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
    "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"
]

# ----------- 目标保存路径 -------------
save_dir = "./graph/Visualized_Sample"
os.makedirs(save_dir, exist_ok=True)

save_path = os.path.join(save_dir, f"fashion_mnist_sample_{idx}.png")
# ---------------------------------------

# 画图
plt.figure(figsize=(4, 4))
plt.imshow(pixels, cmap="gray")
plt.title(f"Label {label}: {label_names[label]}")
plt.axis("off")

# 保存为 PNG
plt.savefig(save_path, dpi=300, bbox_inches='tight')
plt.show()

print(f"图片已保存到: {save_path}")
```
